```python
# Apply KMeans clustering on the banknote.csv dataset. Find the optimal number of clusters
# and then print the clustered dataset.
```

```python
# Imports the dataset and prints the first five rows of the dataset.
import pandas as pd

url = "https://raw.githubusercontent.com/duochen/Python-MachineLearning/master/Reference/PythonMachineLearningforBeginners/Data/banknote.csv"
banknote_df = pd.read_csv(url)
banknote_df.head()
```

```python
# Dividing data into features and labels
features = banknote_df.drop(['class'], axis=1)
labels = banknote_df.filter(['class'], axis=1)
features.head()
```

```python
# Training KMeans on K values from 1 to 10
from sklearn.cluster import KMeans

loss = []
for i in range(1, 11):
    km = KMeans(n_clusters=i).fit(features)
    loss.append(km.inertia_)
```

```python
# Printing loss against number of clusters
import matplotlib.pyplot as plt

plt.plot(range(1, 11), loss)
plt.title('Finding Optimal Clusters via Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('loss')
plt.show()
```

```python
# Training KMeans with 3 clusters
features = features.values
km_model = KMeans(n_clusters=3)
km_model.fit(features)
```

```python
# Printing the data points with predicated labels
plt.scatter(features[:,0], features[:,1], c=km_model.labels_, cmap='rainbow')
```

```python
# Print the predcited centroids
plt.scatter(km_model.cluster_centers_[:,0], km_model.cluster_centers_[:,1], s=100, c='black')
```
